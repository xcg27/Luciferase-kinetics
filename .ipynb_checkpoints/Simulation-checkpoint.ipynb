```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import ipywidgets as widgets
from ipywidgets import interact
from IPython.display import display
```

```python


def enzyme_kinetics(y, t, Vmax, Km, Ki, k_deg):
    E, S, P = y  # E = enzyme, S = substrate, P = product
    
    # Normalized enzyme activity based on substrate inhibition kinetics and enzyme concentration
    v = (Vmax * E * S) / (Km + S + (S**2 / Ki))
    
    # Differential equations
    dE_dt = -k_deg * E  # Enzyme degradation
    dS_dt = -v  # Substrate consumption
    dP_dt = v  # Product formation
    
    return [dE_dt, dS_dt, dP_dt]

def simulate_enzyme_activity(Vmax, Km, Ki, k_deg, E0, S0):
    # Normalize parameters
    max_val = max(Vmax, Km, Ki, k_deg, E0, S0, 1)
    Vmax, Km, Ki, k_deg, E0, S0 = [x / max_val for x in [Vmax, Km, Ki, k_deg, E0, S0]]
    
    # Initial conditions
    P0 = 0.0  # Initial product concentration
    y0 = [E0, S0, P0]
    
    # Time points
    t = np.linspace(0, 50, 500)  # 50 time units with 500 steps
    
    # Solve the system
    solution = odeint(enzyme_kinetics, y0, t, args=(Vmax, Km, Ki, k_deg))
    E_t, S_t, P_t = solution.T
    
    # Compute the reaction rate (production rate) over time with enzyme concentration dependency
    reaction_rate = (Vmax * E_t * S_t) / (Km + S_t + (S_t**2 / Ki))
    
    # Plot results with secondary y-axis for production rate
    fig, ax1 = plt.subplots(figsize=(10, 6))
    ax2 = ax1.twinx()
    
    ax1.plot(t, E_t, label="Enzyme [E]", linestyle="dashed", color='b')
    ax1.plot(t, S_t, label="Substrate [S]", linestyle="dotted", color='g')
    ax2.plot(t, reaction_rate, label="Production Rate (v)", linestyle="solid", color='r')
    
    ax1.set_xlabel("Time")
    ax1.set_ylabel("Normalized Concentration", color='b')
    ax2.set_ylabel("Normalized Production Rate", color='r')
    
    ax1.set_ylim(0, 1)
    ax2.set_ylim(0, 0.2)
    
    ax1.legend(loc='upper left')
    ax2.legend(loc='upper right')
    plt.title("Normalized Enzyme Activity Profile Over Time")
    plt.grid()
    plt.show()

# Create interactive widget
interact(simulate_enzyme_activity,
         Vmax=widgets.FloatSlider(min=0.05, max=1, step=0.05, value=0.5, description='Vmax'),
         Km=widgets.FloatSlider(min=0.05, max=1, step=0.05, value=0.2, description='Km'),
         Ki=widgets.FloatSlider(min=0.05, max=1, step=0.05, value=0.5, description='Ki'),
         k_deg=widgets.FloatSlider(min=0, max=1, step=0.05, value=0.05, description='k_deg'),
         E0=widgets.FloatSlider(min=0.05, max=1, step=0.05, value=0.5, description='E0'),
         S0=widgets.FloatSlider(min=0.05, max=1, step=0.05, value=0.5, description='S0'));
```

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

def michaelis_menten(S, Vmax, Km):
    return (Vmax * S) / (Km + S)

def hill_equation(S, Vmax, Km, n):
    return (Vmax * S**n) / (Km**n + S**n)

def competitive_inhibition(S, Vmax, Km, Ki, I):
    return (Vmax * S) / (Km * (1 + I/Ki) + S)

def uncompetitive_inhibition(S, Vmax, Km, Ki, I):
    return (Vmax * S) / (Km + S * (1 + I/Ki))

def noncompetitive_inhibition(S, Vmax, Km, Ki, I):
    return (Vmax * S) / ((Km + S) * (1 + I/Ki))

def substrate_inhibition(S, Vmax, Km, Ks):
    return (Vmax * S) / (Km + S + (S**2 / Ks))

def update_plot(Vmax, Km, n, Ki, I, Ks):
    S = np.linspace(0, 10, 100)
    rate_mm = michaelis_menten(S, Vmax, Km)
    rate_hill = hill_equation(S, Vmax, Km, n)
    rate_competitive = competitive_inhibition(S, Vmax, Km, Ki, I)
    rate_uncompetitive = uncompetitive_inhibition(S, Vmax, Km, Ki, I)
    rate_noncompetitive = noncompetitive_inhibition(S, Vmax, Km, Ki, I)
    rate_substrate_inhibition = substrate_inhibition(S, Vmax, Km, Ks)

    plt.figure(figsize=(8, 6))
    plt.plot(S, rate_mm, label='Michaelis-Menten', linestyle='-', linewidth=2)
    plt.plot(S, rate_hill, label='Hill Equation (Cooperativity)', linestyle='--', linewidth=2)
    plt.plot(S, rate_competitive, label='Competitive Inhibition', linestyle='-.', linewidth=2)
    plt.plot(S, rate_uncompetitive, label='Uncompetitive Inhibition', linestyle=':', linewidth=2)
    plt.plot(S, rate_noncompetitive, label='Noncompetitive Inhibition', linestyle='-', linewidth=2, alpha=0.7)
    plt.plot(S, rate_substrate_inhibition, label='Substrate Inhibition', linestyle='--', linewidth=2, alpha=0.7)

    plt.xlabel('Substrate Concentration [S]')
    plt.ylabel('Production Rate [v]')
    plt.title('Enzyme Kinetics: Production Rate vs. Substrate Concentration')
    plt.legend()
    plt.grid()
    plt.show()

Vmax_slider = widgets.FloatSlider(value=1.0, min=0.1, max=5.0, step=0.1, description='Vmax')
Km_slider = widgets.FloatSlider(value=2.0, min=0.1, max=5.0, step=0.1, description='Km')
n_slider = widgets.IntSlider(value=2, min=1, max=5, step=1, description='Hill Coeff.')
Ki_slider = widgets.FloatSlider(value=1.0, min=0.1, max=5.0, step=0.1, description='Ki')
I_slider = widgets.FloatSlider(value=1.0, min=0.0, max=5.0, step=0.1, description='Inhibitor [I]')
Ks_slider = widgets.FloatSlider(value=1.0, min=0.1, max=10.0, step=0.1, description='Ks (Substrate Inhibition)')

widgets.interactive(update_plot, Vmax=Vmax_slider, Km=Km_slider, n=n_slider, Ki=Ki_slider, I=I_slider, Ks=Ks_slider)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
import ipywidgets as widgets
from ipywidgets import interact

def simulate_luciferase(
    E0=1.0,
    S0=25.0,
    Km=5.0,
    Ki=200.0,
    kcat=300.0,
    k2=100.0,
    k_2=10.0,
    kdeg=0.01,
    t_end=120
):
    # Base constants
    k1 = 500.0
    k_1 = (Km * k1) - kcat
    ki = 50.0
    k_i = Ki * ki
    E0_uM = E0 / 1000
    
    # Updated ODE system with reversible product step
    def luciferase_model_reversible(t, y):
        S, E, SE, PE, ESS, P = y
        dS_dt = -k1 * E * S + k_1 * SE - ki * SE * S + k_i * ESS
        dE_dt = -k1 * E * S + k_1 * SE + k2 * PE - k_2 * E * P
        dSE_dt = k1 * E * S - (k_1 + kcat) * SE - ki * SE * S + k_i * ESS
        dPE_dt = kcat * SE - k2 * PE + k_2 * E * P - kdeg * PE
        dESS_dt = ki * SE * S - k_i * ESS
        dP_dt = k2 * PE - k_2 * E * P
        return [dS_dt, dE_dt, dSE_dt, dPE_dt, dESS_dt, dP_dt]

    # Time-course simulation
    y0 = [S0, E0_uM, 0.0, 0.0, 0.0, 0.0]
    t_eval = np.linspace(0, t_end, 100)
    sol = solve_ivp(luciferase_model_reversible, (0, t_end), y0, t_eval=t_eval)
    S, E, SE, PE, ESS, P = sol.y

    # Figure settings
    plt.rcParams['font.family'] = 'Verdana'
    plt.rcParams['font.size'] = 12               # base font size
    plt.rcParams['axes.titlesize'] = 16          # title font size
    plt.rcParams['axes.labelsize'] = 12          # axis label size
    plt.rcParams['xtick.labelsize'] = 10         # x tick label size
    plt.rcParams['ytick.labelsize'] = 10         # y tick label size
    plt.rcParams['legend.fontsize'] = 8         # legend size
    plt.rcParams['font.weight'] = 'light'         # regular, bold, heavy, light, etc.
    plt.rcParams['text.color'] = 'black'         # font color

    # Plot results with secondary y-axis for substrate and product
    fig, ax1 = plt.subplots(figsize=(7, 6))
    ax2 = ax1.twinx()
    
    ax2.plot(sol.t, S, label="[S] (substrate)", color='#f58742', linewidth=4, alpha=0.7)
    ax2.plot(sol.t, P, label="[P] (product)", color='#8784c7', linewidth=4, alpha=0.7)
    ax1.plot(sol.t, SE, label="[SE](active enzyme-substrate complex)", color='#00a481', linewidth=5)
    ax1.plot(sol.t, E, label="[E] (free enzyme)", linestyle='-.', color='#42b0f5', linewidth=2, alpha=0.3)
    ax1.plot(sol.t, ESS, label="[ESS] (inhibited complex)", linestyle='--', color='#2b5e9a', linewidth=2, alpha=0.3)
    ax1.plot(sol.t, PE, label="[PE] (post-catalysis complex)", ls=':', color='#465674', linewidth=2, alpha=0.3)

    ax1.set_xlabel("Time", fontweight='bold')
    ax1.set_ylabel("Enzyme Complexes (nM)", color='black', fontweight='bold', labelpad=20, ha='center', va='center')
    ax2.set_ylabel("Substrate (µM)", color='black', fontweight='bold', rotation=270, labelpad=20, ha='center', va='center')

    
    plt.grid(False)
    plt.tight_layout()
    plt.show()

# Interactive widget
interact(simulate_luciferase,
         E0=widgets.FloatLogSlider(base=10, min=-2, max=2, step=0.1, value=1.0, description='[E]₀ (nM)'),
         S0=widgets.FloatSlider(min=25, max=50, step=1, value=25, description='[S]₀ (µM)'),
         Km=widgets.FloatSlider(min=1, max=20, step=1, value=5, description='Km (µM)'),
         Ki=widgets.FloatSlider(min=10, max=1000, step=10, value=200, description='Ki (µM)'),
         kcat=widgets.FloatLogSlider(base=10, min=0, max=3, step=0.1, value=300.0, description='kcat (min⁻¹)'),
         k2=widgets.FloatSlider(min=10, max=500, step=10, value=100.0, description='k2 (min⁻¹)'),
         k_2=widgets.FloatSlider(min=0.0, max=50, step=1, value=10, description='k₋₂ (min⁻¹)'),
         kdeg=widgets.FloatSlider(min=0.0, max=0.5, step=0.01, value=0.0, description='kdeg (min⁻¹)'))

```

## Simulate Vmax decay curve


```python
import numpy as np
import pandas as pd
from scipy.integrate import solve_ivp

def simulate_vmax_vs_time(
    kdeg=0.1,
    Km=10.0,
    Ki=50.0,
    kcat=1.0,
    k2=5.0,
    k_2=0.1,
    E0=1.0,
    S0=25.0
):
    # Base constants
    k1 = 1.0
    ki = 1.0

    # Updated ODE system with reversible product step
    def luciferase_model_reversible(t, y, k_1, k_i, kdeg):
        S, E, SE, PE, ESS, P = y
        dS_dt = -k1 * E * S + k_1 * SE - ki * SE * S + k_i * ESS
        dE_dt = -k1 * E * S + k_1 * SE + k2 * PE - kdeg * E - k_2 * E * P
        dSE_dt = k1 * E * S - (k_1 + kcat) * SE - ki * SE * S + k_i * ESS
        dPE_dt = kcat * SE - k2 * PE + k_2 * E * P
        dESS_dt = ki * SE * S - k_i * ESS
        dP_dt = k2 * PE - k_2 * E * P
        return [dS_dt, dE_dt, dSE_dt, dPE_dt, dESS_dt, dP_dt]

    # Time and substrate settings
    t_points = np.arange(1, 31, 3)
    S_range_dense = np.logspace(np.log10(0.1), np.log10(100), 100)

    # Compute parameters
    k_1 = (Km * k1) - kcat
    k_i = Ki * ki
    vmax_over_time = []

    for t in t_points:
        v_values_at_time = []
        for S_dense in S_range_dense:
            y0_dose = [S_dense, E0, 0.0, 0.0, 0.0, 0.0]
            sol_dose = solve_ivp(luciferase_model_reversible, (0, t + 1), y0_dose, t_eval=[t], args=(k_1, k_i, kdeg))
            SE_val = sol_dose.y[2, 0]
            v_val = kcat * SE_val
            v_values_at_time.append(v_val)
        vmax_over_time.append(np.max(v_values_at_time))

    df = pd.DataFrame({"Time": t_points, f"Vmax_kdeg={kdeg:.2f}": vmax_over_time})
    df.to_csv(f"vmax_vs_time_kdeg_{kdeg:.2f}.csv", index=False)

```
